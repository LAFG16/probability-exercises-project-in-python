## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
# TODO
import pandas as pd
import numpy as np

#Roll 2 dices 1000 times
np.random.seed(24)
dice1 = np.random.randint(1, 7, 1000)
np.random.seed(42)
dice2 = np.random.randint(1, 7, 1000)


both_dices = pd.DataFrame({'dice1': dice1, 'dice2': dice2})
both_dices['total'] = both_dices['dice1'] + both_dices['dice2']

both_dices_track = both_dices[(both_dices['total'] >= 7) | (both_dices['total'] % 2 != 0)]
total_times = both_dices_track['total'].count()

simulation = np.round(total_times/1000 * 100, 2)

print(f'the probability that the total score is even or greater than 7 is: {simulation}%')
```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. All are the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
import random
import numpy as np

random.seed(24)

ball_box = {}

# Create the box of balls
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)
            
# TODO
color_balls = {}

for i in range(1000):
    balls = random.choices(ball_box, k = 5)
    color_balls[i] = balls

for key in color_balls:
    color_balls[key].sort()

same_color = 0
white_red = 0
for colors in color_balls.values():
    if colors == ["White", "White", "White", "White", "White"] or \
        colors == ["Red", "Red", "Red", "Red", "Red"] or \
        colors == ["Green", "Green", "Green", "Green", "Green"]:
        same_color += 1
    elif colors == ["Red", "Red", "White", "White", "White"]:
        white_red +=1

probability1 = same_color/1000

probability2 = white_red/1000

print(f'1) The probability to take 5 balls with same color is: {np.round(probability1*100, 2)}% \
      \n2) The probability to take 3 balls white and 2 balls red is: {np.round(probability2*100, 2)}%')
```
